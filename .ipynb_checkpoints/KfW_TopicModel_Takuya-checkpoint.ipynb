# Preprocessing Text Data

## Preprocessing with the NLTK package

To preprocess the text data, we need to import some packages and download necessary NLTK data.

First, we import the required packages for preprocessing. These packages are:

- `stopwords` from the `nltk.corpus` module: This package provides a list of common words that are often considered irrelevant in text analysis. These words, such as "the", "is", and "and", are commonly known as stopwords and are typically removed from the text before further analysis.

- `WordNetLemmatizer` from the `nltk.stem` module: This package is used for lemmatizing words. Lemmatization is the process of reducing words to their base or dictionary form. For example, the word "running" would be lemmatized to "run".

- `word_tokenize` from the `nltk.tokenize` module: This package is used to tokenize the text into individual words. Tokenization is the process of splitting a text into smaller units, such as words or sentences, for further analysis.

- `re` module: This module provides regular expression matching operations. Regular expressions are powerful tools for pattern matching and manipulation of text data.

- `nltk` module: This is the Natural Language Toolkit library, which provides various tools and resources for natural language processing tasks.

Next, we download the necessary NLTK data. This includes the following:

- `punkt` tokenizer: This tokenizer is used to tokenize text into sentences or words. It is necessary for the `word_tokenize` function.

- `wordnet` lemmatizer: This lemmatizer is used for lemmatizing words to their base form. It is necessary for the `WordNetLemmatizer` function.

- `stopwords` corpus: This corpus contains a list of stopwords that are commonly removed from text during preprocessing. It is necessary for the `stopwords` package.

By importing these packages and downloading the necessary NLTK data, we can perform various preprocessing tasks on text data, such as removing stopwords, lemmatizing words, and tokenizing text.

```python
import nltk
nltk.download('stopwords')  # Download the stopwords corpus
nltk.download('punkt')  # Download the Punkt tokenizer
nltk.download('wordnet')  # Download the WordNet corpus
nltk.download('omw-1.4')  # Download the Open Multilingual WordNet (OMW) corpus

from nltk.corpus import stopwords  # Import the stopwords corpus from NLTK
from nltk.stem import WordNetLemmatizer  # Import the WordNetLemmatizer from NLTK
from nltk.tokenize import word_tokenize  # Import the word_tokenize function from NLTK
import re  # Import the regular expression module

# Set stopwords for German
stop_words = set(stopwords.words('german'))  # Create a set of German stopwords
print(stop_words)  # Print the set of stopwords

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()  # Create an instance of the WordNetLemmatizer

def preprocess_text(text):
    # Remove HTML tags, URLs, digits, and punctuation
    text = re.sub('<.*?>|http\S+|[0-9]+|\W', ' ', text)
    
    # Remove non-words
    text = re.sub(r'\b\w{1,2}\b', '', text)
    
    # Tokenize, lemmatize, and remove stopwords
    text = [lemmatizer.lemmatize(word) for word in word_tokenize(text.lower()) if word not in stop_words]
    
    return ' '.join(text)

# Apply the preprocessing function to the 'description' column
df['description_clean'] = df['description'].apply(preprocess_text)

# Display the first few rows of the DataFrame
df[['description', 'description_clean']].head()
```

After running a first LDA model with the preprocessed text using NLTK, we observed that some stopwords were not removed. Additionally, several frequently-used words including abbreviations, currency units, jargons, and common neutral terms surrounding projects (such as measure, framework, goals, etc.) obscured the topic model analysis. That is why we decided to refine our preprocessing by using the spaCy package for German stopwords and added a list of strings to be removed.

```python
import spacy
import re
import spacy.cli

# Download the German Spacy model
spacy.cli.download("de_core_news_sm")

# Load the German Spacy model
nlp = spacy.load('de_core_news_sm')

def preprocess_de(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove HTML tags, URLs, digits, and punctuation
    text = re.sub('<.*?>|http\S+|[0-9]+|\W', ' ', text)
    # Remove non-words
    text = re.sub(r'\b\w{1,2}\b', '', text)
    # Remove Latin numbers
    text = re.sub(r'\b(?:i{1,3}|iv|v|vi{1,3}|ix|x{1,3}|xl|l|lx{1,3}|xc|c{1,3}|cd|d|dc{1,3}|cm|m{1,3})\b', '', text)
    # Tokenize the text
    doc = nlp(text)
    
    # Define the strings to be removed
    strings_to_remove = ["mio", "eur", "kkmu", "vorhaben", "maßnahme", "ziel", "beitrag", "fz", "rahmen", "phase", "projekt","programm", "modul", "i_h_v", "usd", "somit", "rd", "höhe", "insbesondere", "eac", "bzw"]
    
    # Remove strings containing specific words
    tokens = [token for token in doc if not any(word in token.text.lower() for word in strings_to_remove)]
    
    # Lemmatize the tokens and remove German stopwords
    tokens = [token.lemma_.lower() for token in tokens if not token.is_stop]
    
    # Removing non-words
    tokens = [token for token in tokens if token.isalpha()] 
    
    # Join the tokens back into a string
    processed_text = ' '.join(tokens)
    
    return processed_text
```

```python
# Preprocess project_description using the preprocess_de function
df['description_clean_de'] = df['description'].apply(preprocess_de)

# Display the first few rows of the DataFrame
df[['description', 'description_clean_de']].head()
```

### Build Bigram & Trigram Models

Building bigram and trigram models is useful and helpful for preprocessing German text data for topic models including LDA and BERTtopic. These models capture phrases in the text that frequently appear together, allowing for a more accurate representation of the underlying topics in the data.

In German text, words are often combined to form compound words, and these compound words can carry specific meanings. By using bigram and trigram models, we can identify and preserve these compound words as phrases, rather than treating them as separate individual words. This helps to maintain the semantic context and improve the accuracy of topic modeling.

For example, in the context of German text, the words "Klima" (climate) and "Wandel" (change) might often appear together as the compound word "Klimawandel" (climate change). By building bigram and trigram models, we can identify and treat "Klimawandel" as a single phrase, which is more meaningful and informative for topic modeling.

In summary, building bigram and trigram models for German text data preprocessing enhances the accuracy and effectiveness of topic models like LDA and BERTtopic by capturing meaningful compound words and preserving their semantic context.

```python
import spacy
from gensim.models import Phrases

# Load spaCy's German model
nlp = spacy.load("de_core_news_sm")

# Tokenize and lemmatize with spaCy
def tokenize_and_lemmatize(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]

# Apply the function
df['lemmatized_descriptions'] = df['description_clean_de'].apply(tokenize_and_lemmatize)

# Build bigram and trigram models
bigram = Phrases(df['lemmatized_descriptions'].tolist(), min_count=5, threshold=100)
trigram = Phrases(bigram[df['lemmatized_descriptions'].tolist()], threshold=100)

# Function to get bigrams and trigrams
def get_ngrams(text):
    bigrams_ = bigram[text]
    trigrams_ = trigram[bigram[text]]
    return trigrams_

# Apply function to the descriptions
df['final_descriptions'] = df['lemmatized_descriptions'].apply(get_ngrams)
```

# Topic Modeling

## Gensin for Topic Modeling

Gensim is a Python library for topic modeling and document similarity analysis. It provides an easy-to-use interface for training and using topic models such as Latent Dirichlet Allocation (LDA) and Latent Semantic Analysis (LSA). Gensim also offers functionality for word embeddings, document indexing, and similarity retrieval.

LDA topic modeling is often preferred over LSA for several reasons:

1. LDA is a generative probabilistic model, which means it assumes that documents are generated from a mixture of topics. This allows LDA to capture the underlying thematic structure of the documents more effectively.

2. LDA provides a more interpretable representation of topics. Each topic is represented as a distribution over words, and each document is represented as a distribution over topics. This makes it easier to understand and analyze the topics extracted by the model.

3. LDA allows for topic inference on new, unseen documents. Once the LDA model is trained, it can be used to infer the topic distribution of new documents, even if they were not part of the training data. This makes LDA more suitable for real-time applications where new documents need to be analyzed.

In this research project, we will use Gensim to perform topic modeling on the preprocessed project descriptions using LDA.

## Creating Dictionary and Corpus

In Gensim, a dictionary is a mapping between words and their integer ids. To create a dictionary from your text data, you can use the `corpora.Dictionary` class. 

A corpus, on the other hand, is a collection of documents represented as a stream of vectors. Each vector represents a document and contains the word frequencies or the word ids and their corresponding frequencies.

In the code chunk below, `documents` is a list of lists, where each inner list represents a document and contains the words in that document. The `doc2bow` method converts each document into a bag-of-words representation.

Training the LDA (Latent Dirichlet Allocation) model involves fitting the model to a given corpus of documents. The purpose training the LDA model is to discover the underlying topics within the corpus and assign topic probabilities to each document.

To train the LDA model, we first need to create a dictionary and corpus. The dictionary maps words to their integer ids, while the corpus represents the documents as a stream of vectors, where each vector contains the word frequencies or word ids and their corresponding frequencies.

Once the dictionary and corpus are created, we can proceed with training the LDA model. The LDA algorithm iteratively assigns topics to words and documents, aiming to maximize the likelihood of the observed data. This process involves estimating the topic-word and document-topic distributions.

By training the LDA model, we can gain insights into the main themes or topics present in the corpus. This can be useful for various applications such as document classification, information retrieval, and topic summarization.

Training the LDA model allows us to uncover the latent structure of the corpus and extract meaningful information from the text data. It helps in understanding the relationships between words and documents, and provides a way to organize and analyze large collections of textual data.

```python
# Perform LDA topic modeling
from gensim import corpora, models

# Create dictionary
dictionary = corpora.Dictionary(df['description_clean'].apply(lambda x: x.split()).tolist())

# Create corpus
corpus = [dictionary.doc2bow(text.split()) for text in df['description_clean']]
```

```python
# Train LDA model
lda_model = models.LdaModel(corpus=corpus, num_topics=10, id2word=dictionary, passes=10)

# Show top 10 frequent words of each topic
for topic_id in range(lda_model.num_topics):
    topic_words = lda_model.show_topic(topic_id, topn=20)
    print(f"Topic {topic_id + 1}:")
    for word, freq in topic_words:
        print(f"{word}: {freq}")
    print()

```

## Topic Modeling with CountVectorizer
To initialize the CountVectorizer and vectorize the clean_description, the following steps are taken:

1. Import the CountVectorizer class from the sklearn.feature_extraction.text module.

2. Create an instance of the CountVectorizer class.

3. Set the desired parameters for the CountVectorizer, such as the maximum number of features, the minimum document frequency, and the n-gram range.

4. Fit the CountVectorizer to the clean_description data, which learns the vocabulary from the text data.

5. Transform the clean_description data into a matrix of token counts using the transform method of the CountVectorizer.

6. The resulting matrix represents the vectorized representation of the clean_description data, where each row corresponds to a document and each column corresponds to a unique token in the vocabulary.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Initialize CountVectorizer
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words= stop_words)

# Vectorize the cleaned description
data_vectorized = vectorizer.fit_transform(df['description_clean_de'])
data_vectorized
```

The next code chunk performs topic modeling using Latent Dirichlet Allocation (LDA) on the vectorized description data.

1. Import the LatentDirichletAllocation class from the sklearn.decomposition module.

2. Initialize an instance of the LatentDirichletAllocation class with the desired number of topics.

3. Fit the LDA model to the vectorized description data using the fit_transform method.

4. The resulting model represents the learned topics, where each topic is a distribution over the vocabulary.

5. The transform method can be used to infer the topic distribution for new documents.

6. The components_ attribute of the LDA model contains the topic-word distributions, where each row corresponds to a topic and each column corresponds to a word in the vocabulary.

```python
# Initialize LDA Model with 10 topics
lda_model = LatentDirichletAllocation(n_components=10, random_state=0)

# Fit the model on the vectorized data
lda_model.fit(data_vectorized)

# Visualize the bar chart of the top words for each topic
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx + 1}:")
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
display_topics(lda_model, vectorizer.get_feature_names_out(), no_top_words)
```

The code above applies Latent Dirichlet Allocation (LDA) topic modeling on the vectorized description data. LDA is a generative probabilistic model that represents documents as mixtures of topics. Each topic is a distribution over words, and each document is assumed to be a mixture of topics. The LDA model is trained on the vectorized data using the fit method.

The display_topics function is defined to visualize the top words for each topic. It iterates over the components_ attribute of the LDA model, which contains the topic-word distributions. For each topic, it prints the topic index and the top words based on their probabilities in the topic.

The output of the code is a bar chart of the top words for each topic, with each topic labeled by its index. The number of top words to display is specified by the no_top_words variable. The top words are determined by sorting the topic-word distribution in descending order and selecting the top words.

By analyzing the top words for each topic, we can gain insights into the main themes or subjects present in the description data. These topics can be used to categorize or classify the documents, or to understand the underlying patterns in the data.

```python
import matplotlib.pyplot as plt

def barchart_topics(model, feature_names, no_top_words):
    # Create a figure and an array of subplots with the specified number of rows and size
    fig, axes = plt.subplots(nrows=model.n_components, figsize=(10, 10), sharex=True)
    
    # Get the topic-word distributions from the model
    topics = model.components_

    # Iterate over each topic
    for topic_idx, topic in enumerate(topics):
        # Get the indices of the top words for the current topic
        top_words_idx = topic.argsort()[:-no_top_words - 1:-1]
        
        # Get the actual top words and their corresponding scores
        top_words = [feature_names[i] for i in top_words_idx]
        top_scores = [topic[i] for i in top_words_idx]

        # Get the subplot for the current topic
        ax = axes[topic_idx]
        
        # Create a horizontal bar chart with the top words and their scores
        ax.barh(top_words, top_scores)
        
        # Set the title of the subplot to indicate the topic number
        ax.set_title(f"Topic {topic_idx + 1}")
        
        # Invert the y-axis to display the top words at the top
        ax.invert_yaxis()
        
        # Set the label size of the y-axis ticks to 8
        ax.tick_params(axis='y', labelsize=8)

    # Adjust the layout of the subplots to prevent overlapping
    plt.tight_layout()
    
    # Display the plot
    plt.show()

# Specify the number of top words to display for each topic
no_top_words = 8

# Call the barchart_topics function with the specified arguments
barchart_topics(lda_model, vectorizer.get_feature_names_out(), no_top_words)
```

```python
# Analyzing Term Frequency of Top Words

def analyze_term_frequency(lda_model, num_words=10):
    # Get the top words for each topic
    top_words = lda_model.show_topics(num_topics=-1, num_words=num_words, formatted=False)
    
    # Plot the term frequency of top words for each topic
    plt.figure(figsize=(12, 8))
    for topic_id, words in top_words:
        word_list = [word[0] for word in words]
        word_freq = [word[1] for word in words]
        plt.bar(word_list, word_freq, alpha=0.7, label=f'Topic {topic_id}')
    plt.xlabel('Top Words')
    plt.ylabel('Term Frequency')
    plt.title('Term Frequency of Top Words in Each Topic')
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()

# Call the function
analyze_term_frequency(lda_model)
```
